```python
dataset_path = "data.csv"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load the dataset
data = pd.read_csv(dataset_path)
```

```python
# # Normalizing the metrics
try: 
    data['Normalized_MarianMT_BLEU'] = (data['MarianMT_BLEU'] - data['MarianMT_BLEU'].min()) / (data['MarianMT_BLEU'].max() - data['MarianMT_BLEU'].min())
    data['Normalized_mBART_BLEU'] = (data['mBART_BLEU'] - data['mBART_BLEU'].min()) / (data['mBART_BLEU'].max() - data['mBART_BLEU'].min())
    data['Normalized_Tsmall100_BLEU'] = (data['Tsmall100_BLEU'] - data['Tsmall100_BLEU'].min()) / (data['Tsmall100_BLEU'].max() - data['Tsmall100_BLEU'].min())

    data['Normalized_MarianMT_TER'] = 1 - (data['MarianMT_TER'] - data['MarianMT_TER'].min()) / (data['MarianMT_TER'].max() - data['MarianMT_TER'].min())
    data['Normalized_mBART_TER'] = 1 - (data['mBART_TER'] - data['mBART_TER'].min()) / (data['mBART_TER'].max() - data['mBART_TER'].min())
    data['Normalized_Tsmall100_TER'] = 1 - (data['Tsmall100_TER'] - data['Tsmall100_TER'].min()) / (data['Tsmall100_TER'].max() - data['Tsmall100_TER'].min())

    data['Normalized_MarianMT_CHRF'] = (data['MarianMT_CHRF'] - data['MarianMT_CHRF'].min()) / (data['MarianMT_CHRF'].max() - data['MarianMT_CHRF'].min())
    data['Normalized_mBART_CHRF'] = (data['mBART_CHRF'] - data['mBART_CHRF'].min()) / (data['mBART_CHRF'].max() - data['mBART_CHRF'].min())
    data['Normalized_Tsmall100_CHRF'] = (data['Tsmall100_CHRF'] - data['Tsmall100_CHRF'].min()) / (data['Tsmall100_CHRF'].max() - data['Tsmall100_CHRF'].min())

    data['Normalized_MarianMT_ROUGE'] = (data['MarianMT_ROUGE'] - data['MarianMT_ROUGE'].min()) / (data['MarianMT_ROUGE'].max() - data['MarianMT_ROUGE'].min())
    data['Normalized_mBART_ROUGE'] = (data['mBART_ROUGE'] - data['mBART_ROUGE'].min()) / (data['mBART_ROUGE'].max() - data['mBART_ROUGE'].min())
    data['Normalized_Tsmall100_ROUGE'] = (data['Tsmall100_ROUGE'] - data['Tsmall100_ROUGE'].min()) / (data['Tsmall100_ROUGE'].max() - data['Tsmall100_ROUGE'].min())

    data['Normalized_MarianMT_COMET'] = (data['MarianMT_COMET'] - data['MarianMT_COMET'].min()) / (data['MarianMT_COMET'].max() - data['MarianMT_COMET'].min())
    data['Normalized_mBART_COMET'] = (data['mBART_COMET'] - data['mBART_COMET'].min()) / (data['mBART_COMET'].max() - data['mBART_COMET'].min())
    data['Normalized_Tsmall100_COMET'] = (data['Tsmall100_COMET'] - data['Tsmall100_COMET'].min()) / (data['Tsmall100_COMET'].max() - data['Tsmall100_COMET'].min())

    data['Normalized_MarianMT_BLURT'] = (data['MarianMT_BLURT'] - data['MarianMT_BLURT'].min()) / (data['MarianMT_BLURT'].max() - data['MarianMT_BLURT'].min())
    data['Normalized_mBART_BLURT'] = (data['mBART_BLURT'] - data['mBART_BLURT'].min()) / (data['mBART_BLURT'].max() - data['mBART_BLURT'].min())
    data['Normalized_Tsmall100_BLURT'] = (data['Tsmall100_BLURT'] - data['Tsmall100_BLURT'].min()) / (data['Tsmall100_BLURT'].max() - data['Tsmall100_BLURT'].min())

    # Combine the normalized metrics (assuming equal weightage)
    data['Final_Quantitative_Score_MarianMT'] = (
        data['Normalized_MarianMT_BLEU'] + data['Normalized_MarianMT_TER'] + 
        data['Normalized_MarianMT_CHRF'] + data['Normalized_MarianMT_ROUGE'] +
        data['Normalized_MarianMT_COMET'] + data['Normalized_MarianMT_BLURT']) / 6

    data['Final_Quantitative_Score_mBART'] = (
        data['Normalized_mBART_BLEU'] + data['Normalized_mBART_TER'] + 
        data['Normalized_mBART_CHRF'] + data['Normalized_mBART_ROUGE'] +
        data['Normalized_mBART_COMET'] + data['Normalized_mBART_BLURT']) / 6

    data['Final_Quantitative_Score_Tsmall100'] = (
        data['Normalized_Tsmall100_BLEU'] + data['Normalized_Tsmall100_TER'] + 
        data['Normalized_Tsmall100_CHRF'] + data['Normalized_Tsmall100_ROUGE'] +
        data['Normalized_Tsmall100_COMET'] + data['Normalized_Tsmall100_BLURT']) / 6
except:
    pass
```
